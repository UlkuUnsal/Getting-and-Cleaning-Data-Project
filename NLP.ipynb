```python
#reading data and pre-processing

import os
import pandas as pd

 # get the files from bbc website
 # download and unzip all the files
!wget http://mlg.ucd.ie/files/datasets/bbc-fulltext.zip
!unzip bbc-fulltext.zip

#path for bbc data
data_dir = "bbc"

# eliminate for readme.txt
# find all categories
categories = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]

# empty list for saving files
data = []

for category in categories:
    category_path = os.path.join(data_dir, category)       # full path for categories
    for filename in os.listdir(category_path):
        filepath = os.path.join(category_path, filename)   # full path for .txt files
        with open(filepath, 'r', encoding='latin1') as f:
            content = f.read()
            data.append({                                  # assign a dictionaty for all categories and texts
                "category": category,
                "text": content
            })

df = pd.DataFrame(data)

df.head()
```
